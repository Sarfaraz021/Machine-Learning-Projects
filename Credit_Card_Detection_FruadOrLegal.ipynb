# Under-Sampling
Build a sample dataset containing similar distribution of normal 

transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample=legit.sample(n=492)
legit_sample.head()
```

```python
legit_sample.describe()
```

```python
legit_sample.shape
```

# Concatenating two DataFrames

```python
credit_card_dataset=pd.concat([fraud,legit_sample],axis=0)
```

```python
credit_card_dataset.head()
```

```python
credit_card_dataset.Class.value_counts
```

# HERE WE COMPLETE THE DATA PREPROCESSING STEP  NOW SPLIT THE DATA INTO INTO TRAIN AND TEST DATASET

```python
Inputs=credit_card_dataset.drop(['Class'],axis=1)
```

```python
Target=credit_card_dataset.Class
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(Inputs,Target,test_size=25,random_state=10)
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import  LogisticRegression
Lg_model=LogisticRegression()
```

```python
Lg_model.fit(x_train,y_train)
```

```python
Lg_model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# accuracy on test data
x_test_pred=Lg_model.predict(x_test)
Acc=accuracy_score(x_test_pred,y_test)
print('Accuracy score of model is :)' , Acc )
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(x_test_pred,y_test)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predictive classes')
plt.ylabel('Truth Classes')
```
